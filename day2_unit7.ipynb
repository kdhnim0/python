```python
import csv
import matplotlib.pyplot as plt
f=open('age.csv')
data=csv.reader(f)
result=[]
dong = input("인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해 주세요 : ")
for row in data:
    if dong in row[0].split('(')[0].split(' ')[-1]: #동만 비교할수 있도록 구분하여 비교 
        for i in row[3:]:
            #print(i.replace(',',''))
            result.append(int(i.replace(',',''))) # csv의 천단위이상의 ','를 제거
if result==[]:
    print("해당하는 동이 없습니다.")
else:
    plt.style.use('ggplot') # 격자 무늬 스타일 지정
    plt.plot(result)
    plt.show()
f.close()
```
