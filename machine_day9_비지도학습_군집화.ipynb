```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.randint(0,10,(100,2)).astype(np.float32)

x.shape
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler, Normalizer
s0 = MinMaxScaler().fit(x) # 최소값, 최대값 기억해 둔다
x0 = s0.transform(x) # 기억해둔 최소값 최대값으로 정규화한다
```

```python
plt.scatter(x[:,0], x[:,1])
plt.scatter(x0[:,0], x0[:,1])
plt.xlim(-10,10)
plt.ylim(-10,10)
```

```python
s1 = StandardScaler().fit(x) # 평균과 표준편차를 기억해둔다.
x1 = s1.transform(x)
```

```python
plt.scatter(x[:,0], x[:,1])
plt.scatter(x1[:,0], x1[:,1])
plt.xlim(-10,10)
plt.ylim(-10,10)
```

```python
s2 = RobustScaler().fit(x)
x2 = s2.transform(x)
s3 = Normalizer().fit(x)
x3 = s3.transform(x)
```

```python
plt.scatter(x[:,0], x[:,1])
plt.scatter(x2[:,0], x2[:,1])
plt.scatter(x3[:,0], x3[:,1])
plt.xlim(-10,10)
plt.ylim(-10,10)
```

```python
np.sum(s0.inverse_transform(x0) - x)#역변환
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
# 유방암 데이터를 2차원으로 주성분을 뽑아서 그래프로 표현합니다.
cancer = load_breast_cancer()
cancer.data.shape # 30차원
```

```python
scalar = StandardScaler().fit(cancer.data) #평균과 표준편차를 계산해 봅니다.
```

```python
x = scalar.transform(cancer.data)
np.max(cancer.data), np.max(x)
```

```python
pca = PCA(2).fit(x)#30차원을 2차원으로 축소 (정보의 손실을 최소화하면서 데이터를 축소)
```

```python
x_pca = pca.transform(x)
x_pca.shape 
```
