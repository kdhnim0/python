```python
combined_news = combined_news.dropna()
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
X = combined_news.drop(columns='Label')
y = combined_news['Label']
X.shape
```

```python
X_flat = X.values.reshape([-1])
X_flat.shape
```

```python
#vect = CountVectorizer().fit(X_flat)
#vect = TfidfVectorizer(ngram_range=(1, 2)).fit(X_flat)
vect = TfidfVectorizer(min_df=0.03, max_df=0.97, ngram_range=(2, 2), max_features=200000).fit(X_flat)
# 3%이하 / 97% 이상 등장하는 단어는 제외, 최대어휘 20만 제한, N그랩 2(단어 2개씩 자르기)
```

```python
vect.vocabulary_
```

```python
sm= vect.transform(X_flat)
```

```python
sm.toarray(), sm, len(y)
```

```python
train_m = sum(combined_news.index <= '2014-12-31')
train_m
```

```python
feature_n = X.shape[1]
```

```python
y_train = y.values[:train_m]
y_test = y.values[train_m:]
```

선형모델, X의 dim이 2이어야 합니다.
> 방안1. : 뉴스1~25를 합한다
> 방안2. : spares_matrix 25개를 합친다(더한다)

```python
x_agr = []
for i in range(len(X)):
    one_day_news = sm[i*25 : (i+1)*25] # <25x34682>
    new_sum = np.sum(one_day_news, axis = 0) # <34682>
    x_agr.append(new_sum)
len(x_agr)
```

```python
x_2d = np.concatenate(x_agr, axis=0)
```

```python
x_2d.shape
```

```python
x_train = x_2d[:train_m] # 1608 * 25개의 뉴스
x_test = x_2d[train_m:]
x_train.shape, y_train.shape
```

```python
from sklearn.linear_model import LogisticRegression, RidgeClassifier
```

```python
model = LogisticRegression().fit(x_train, y_train)
```

```python
model.score(x_test,y_test)
```

```python
md = RidgeClassifier().fit(x_train, y_train)
md.score(x_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
score = cross_val_score(LogisticRegression(), x_2d, y.values, cv=7)
score
```
